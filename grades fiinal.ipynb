## I am going to try and guess the gender of the person by the other atributes.

#### the stages of how to classify:
##### stage 1: split the data into 2 part train and set.
##### stage 2: divide the splitfata into x,y. x is all of the paramitures with which we will try to guess the class. y in the class.
#####  stage 3: fit the model with the train x and y.
#####  stage 4: redict the classes of the x test. and compare to actual y test.
#####  stage 5: cross validation.
checking are result by checking are model on smaller groups and making sure that we didnt overfit the model
##### stage 6: confussion metrix 
Gives us information about how correct are model is and where the classification mistakes are.
<br> true negitive(top left) true female
<br> false positive(top right) female classified as male
<br> false negative(bottom right) male classified as female
<br> true positive(bottom left) true male
#####  stage 7: precision score, recall score and f1 score.
percision score-true positive/(true positive + false positive)= true positive/(predicted positive)
<br> recall score-true positive/(true positive + false negative)=true positive/(actually positive)
<br> f1 score-2 * ((percision * recall)/(precision + recall))
#####  stage 8: roc curve, roc auc score
roc curve is the coralation between tpr- y axis, fpr- x axis.
<br> tpr- true positive rate = recall
<br> fpr- false positive rate = false positive/(false positive/true negitive)
<br> roc auc score= the area under the roc curve


In my projet I am trying to predict the gender so the most importantent thing is the the cross validation score and how well the model works. 
<br> Not how high the recall score is, f1 score is and how high the roc auc score.
<br> So I will show how to do steps 6-8 for 1 model and not for the rest because it isnt relevent for what i am trying to classify.

## knn
knn classifies by looking at all of the neighbors and seeing which class has more neighbors and by that classifying.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
count1=0
for x in range (2,50):
    knn = KNeighborsClassifier(n_neighbors=x)
    knn.fit(X_Train_set1, Y_Train_set1)
    y_pred2 = knn.predict(X_test_set1)
    a2=metrics.accuracy_score(Y_test_set1, y_pred2)
    if a2>count1:
        count1=a2
        print(x , count1)
```

finding which knn we should use, seeing with how many neighbors do we get the bset result. here it is 45 neighbors.

```python
knn = KNeighborsClassifier(n_neighbors=45)
knn.fit(X_Train_set1, Y_Train_set1)
y_pred3 = knn.predict(X_test_set1)
a3=metrics.accuracy_score(Y_test_set1, y_pred3)
print(a3)
```
