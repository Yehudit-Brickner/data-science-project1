## Linear Regression
looks at the data and creates a linear graoh where the distance squared to the line from each sample is the least, and it maps its result on this line.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_Train_set1, Y_Train_set1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
lin_reg.predict(X_test_set1)[0:5]
```

```python
Y_test_set1[0:5]
```

```python
from sklearn.metrics import mean_squared_error
cars_predictions_linear1 = lin_reg.predict(X_test_set1)
lin_mse = mean_squared_error(Y_test_set1, cars_predictions_linear1) # mean squared error
lin_rmse = np.sqrt(lin_mse) #square root of mean squared error
lin_rmse
```

r2= 1-(((ytrue-ypred)*ytrue-ypred)).sum()/((ytrue-ytrue.mean())*(ytrue-ytrue.mean())).sum())
<br> mean absolute error= 1/(the amount of sampels)* sigma (from i=0  to i=n-1) (|yi actual-yi pred|)
<br> mean squared error= 1/(the amount of sampels)* sigma (from i=0  to i=n-1) (yi actual-yi pred)*(i actual-yi pred)

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(r2_score(cars_predictions_linear1,Y_test_set1))
print(mean_absolute_error(cars_predictions_linear1,Y_test_set1))
print(mean_squared_error(cars_predictions_linear1,Y_test_set1))
```

```python
plt.figure(figsize=(6,6))
plt.scatter(cars_predictions_linear1,Y_test_set1)
```

in this graph we can see that there the prediction is very similar to the actual value

cross validation-
checking are result by checking are model on smaller groups and making sure that we didnt overfit the model.

```python
from sklearn.model_selection import cross_val_score

scores_linear = cross_val_score(lin_reg, X_Train_set1, Y_Train_set1, scoring="neg_mean_squared_error", cv=5) # returns negitive value of mean squared error
lin_rmse_scores = np.sqrt(-scores_linear) # because the number is negitive we will times it by -1, so that we can do the square root
print(lin_rmse_scores, "mean:",np.mean(lin_rmse_scores))
```

lin_rmse_scores: pre cross validation ~8998 after cross validation the average is ~8743 . the model is as good as we thogut or better.
